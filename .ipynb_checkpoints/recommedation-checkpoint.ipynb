popularity based recommender system

```python
rating_with_name = rating.merge(book, on="ISBN")
```

```python
num_rate_df = rating_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rate_df.rename(columns={"Book-Rating":"num-ratings"}, inplace= True)
num_rate_df
```

```python
rating_with_name['Book-Rating'] = pd.to_numeric(rating_with_name['Book-Rating'], errors='coerce')
rating_with_name.dropna(subset=['Book-Rating'], inplace=True)
```

```python
# Drop non-numeric columns except 'Book-Title'
numeric_columns = rating_with_name.select_dtypes(include=['number']).columns
rating_with_name_numeric = rating_with_name[["Book-Title"] + list(numeric_columns)]

# Group by 'Book-Title' and calculate the median for 'Book-Rating'
avg_rate_df = rating_with_name_numeric.groupby("Book-Title").mean()["Book-Rating"].reset_index()

# Rename the 'Book-Rating' column to 'avg-ratings'
avg_rate_df.rename(columns={"Book-Rating": "avg-ratings"}, inplace=True)

avg_rate_df
```

```python
popular_df= num_rate_df.merge(avg_rate_df, on="Book-Title")
print(popular_df)
```

```python
popular_df[popular_df['num-ratings'] >=250].sort_values("avg-ratings", ascending= False).head(50)
```

```python
# Merge the DataFrames, drop duplicates based on 'Book-Title', and select the required columns
result_df = (
    popular_df
    .merge(book, on='Book-Title')
    .drop_duplicates('Book-Title')
    [['Book-Title', 'Book-Author', 'Image-URL-M', 'num-ratings', 'avg-ratings']]
)

result_df
```

```python
##COLLABORATIVE FILTERING BASED RECOMMENDER SYSTEM

x=rating_with_name.groupby('User-ID').count()['Book-Rating'] >=50
educated_users =x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(educated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()["Book-Rating"] >=20
famous_books= y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index= "Book-Title", columns= "User-ID", values="Book-Rating")
```

```python
pt.fillna(0, inplace= True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores= cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    if book_name not in pt.index:
        return f"'{book_name}' not found in the index."

    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)

    data = []
    for i in similar_items:
        item = []
        temp_df = book[book['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Author')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Image-URL-M')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('Whispers')
```
